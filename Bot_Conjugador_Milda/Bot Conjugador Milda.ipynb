El bot propone un juego/test de preguntas y respuestas en el que el usuario ha de escribir la forma verbal que se le pide. El bot evalúa la respuesta y muestra la respuesta correcta. El bot no tiene muchas funciones porque la parte verdaderamente complicada es la creación y validación del corpus del que se extraen las palabras.

```python
import telebot
import random

preguntas_realizadas = 0
preguntas_acertadas = 0
respuesta_ultima_pregunta = ""

# credenciales
TOKEN = '258581581:AAHqBOAWdXWi8eTOlzDmAgkQqNVU5JR11o8' # Milda
bot = telebot.TeleBot(TOKEN)

def forma_verbal_aleatoria_aux(terminaciones, ruta):
    """Creamos una función recursiva que recorre el árbol de las terminaciones posibles y 
    elige una clave aleatoriamente en cada nivel del árbol. Una vez elegida la clave, 
    la añade a la lista 'ruta'."""
    clave = random.choice(list(terminaciones.keys()))

    # La conjugación no es parte necesaria de la ruta, por lo que al llegar a ese nivel, paramos.
    if clave in ["C1", "C2", "C3"]:
        return ruta
    
    ruta.append(clave)
    
    # Si aún hay más árbol que recorrer, volvemos a llamar a la función
    if isinstance(terminaciones[clave], dict):
        return forma_verbal_aleatoria_aux(terminaciones[clave], ruta)
    else:
        return ruta


def forma_verbal_aleatoria(terminaciones):
    """Devuelve una lista que representa una ruta aleatoria en el árbol de terminaciones."""
    return forma_verbal_aleatoria_aux(terminaciones, [])


def convierte_ruta_en_cadena(ruta):
    trozos_cadena = []
    for nivel in ruta:
        # "formas_flexivas_v", "formas_invariables_v" y "participio" son categorías
        # que incluimos en el árbol por comodidad, pero no forman parte de la descripción
        # verbal, por lo que debemos excluirlas.
        if nivel in ("formas_flexivas_v", "formas_invariables_v", "participio"):
            pass
        else:
            nivel = nivel.replace("_", " ")
            trozos_cadena.append(nivel)
            
    return " > ".join(trozos_cadena)


def busca_terminacion(ruta_forma_verbal, conjugacion):
    """Busca la terminación que corresponde a la forma verbal en cuestión"""
    diccionario = morfologia_v
    for nivel in ruta_forma_verbal:
        diccionario = diccionario[nivel]       
    if "C1" in diccionario:
        return diccionario[conjugacion]
    else:
        return diccionario

# En lituano unas formas se construyen a partir del tronco (palabra menos la terminación)
# del infinitivo, otras a partir del tronco de la 3PSgPr y otras de la 3PSgPs.  
def se_construye_a_partir_del_pasado(ruta_forma_verbal):
    if ruta_forma_verbal[0] == "formas_flexivas_v" and ruta_forma_verbal[1] == "pasado":
        resultado = True
    elif ruta_forma_verbal[0] == "formas_invariables_v" and ruta_forma_verbal[1] == "gerundio" and ruta_forma_verbal[2] == "pasado":
        resultado = True
    else:
        resultado = False

    return resultado

def se_construye_a_partir_del_presente(ruta_forma_verbal):
    if ruta_forma_verbal[0] == "formas_flexivas_v" and ruta_forma_verbal[1] == "presente":
        resultado = True
    elif ruta_forma_verbal[0] == "formas_invariables_v" and ruta_forma_verbal[1] == "gerundio" and ruta_forma_verbal[2] == "presente":
        resultado = True
    else:
        resultado = False

    return resultado

def conjuga_verbo(verbo, ruta_forma_verbal):
    """Determina la conjugación del verbo a partir de la 3PSgPr."""
    presente= verbo[1]
    if presente[-1] == "a":
        conjugacion = "C1"
    elif presente[-1] == "i":
        conjugacion = "C2"
    elif presente[-1] == "o":
        conjugacion = "C3"
        
    terminacion = busca_terminacion(ruta_forma_verbal, conjugacion)
    
    if se_construye_a_partir_del_pasado(ruta_forma_verbal):
        # Eliminamos la última letra del la 3PSgPs para obtener el tronco a partir del cual
        # construiremos la conjugación
        tronco = verbo[2][:-1]
        # En este caso particular, hay que eliminar una letra adicional.
        if ruta_forma_verbal[0] == "formas_flexivas_v" and \
           ruta_forma_verbal[1] == "pasado" and \
           ruta_forma_verbal[2] == "participio" and \
           ruta_forma_verbal[3] == "participio_pasivo" and \
           tronco[:-1] == "j":
            tronco = tronco[:-1]
        
    elif se_construye_a_partir_del_presente(ruta_forma_verbal):
        tronco = verbo[1][:-1]
    else:
        tronco = verbo[0][:-2]
    
    return tronco + terminacion


# función para manejar el comando /start
@bot.message_handler(commands=['start'])
def enviar_instrucciones(message):
    chat_id = message.chat.id
    bot.send_message(chat_id, """
    Labas! 
    Yo soy un bot medio lituano que te ayudará a practicar la conjugación de los verbos lituanos. Sé que pueden ser difíciles de aprender pero no te preocupes, puedes repetir las veces que quieras.
    Para pedirme un test de diez verbos escribe /jugar. Cuando termines, te diré cuántos has hecho bien. ¡Suerte!
    """)

# función para manejar el comando /jugar
@bot.message_handler(commands=['jugar'])
def conjugar_verbos(message):
    global preguntas_realizadas
    global respuesta_ultima_pregunta

    chat_id = message.chat.id
    
    preguntas_realizadas = preguntas_realizadas + 1
    
    # Cogemos un verbo aleatorio de la lista del corpus
    verbo = random.choice(lista_verbos)
    ruta_forma_verbal = forma_verbal_aleatoria(morfologia_v)
    
    respuesta_ultima_pregunta = conjuga_verbo(verbo, ruta_forma_verbal)
    
    ruta_texto = convierte_ruta_en_cadena(ruta_forma_verbal)
    
    bot.send_message(chat_id, """
    ¡Empezamos el juego! Aquí va el verbo:
    Escríbeme {} del verbo "{}"
    """.format(ruta_texto, verbo[0], respuesta_ultima_pregunta))
    
@bot.message_handler(func=lambda message: True, content_types=["text"])
def command_default(message):
    global preguntas_realizadas
    global preguntas_acertadas
    global respuesta_ultima_pregunta
    
    chat_id = message.chat.id
    
    #Recogemos lo que responde el usuario 
    respuesta_usuario = message.text.strip()
    
    mensaje = ""
    
    # Comprobamos si la respuesta es válida
    if respuesta_usuario == respuesta_ultima_pregunta:
        preguntas_acertadas += 1
        
        mensaje = "¡Correcto!"
    else:
        mensaje = "¡Error! La respuesta correcta era \"{}\"".format(respuesta_ultima_pregunta)
        
    mensaje += """
    Puntuación: {}/{}
    """.format(preguntas_acertadas, preguntas_realizadas)
    
    # Si el usuario ya ha contestado 10 preguntas seguidas, reiniciamos el marcador.
    if preguntas_realizadas == 10:
        preguntas_acertadas = 0
        preguntas_realizadas = 0
        mensaje += "¡Has terminado la ronda! Puedes empezar de nuevo escribiendo /jugar"
    else:
        mensaje += "Para jugar otra vez, escribe /jugar"
    
    bot.send_message(chat_id, mensaje)
    

bot.polling()
```
